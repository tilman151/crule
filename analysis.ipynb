```python
import importlib

import numpy as np

from crule import evaluation

evaluation = importlib.reload(evaluation)
from crule.evaluation import load_runs
```

```python
runs = load_runs("rul-adapt/benchmark", exclude_tags=["pretraining", "full_eval_do_not_use", "unstable", "eval_degraded_only", "extended"], with_epoch_time=True)
runs.groupby("approach").count()
```

```python
def scrub_strings(metric, df):
    types = df[metric].apply(lambda x: type(x))
    df = df.drop(index=df.iloc[np.argwhere(types == str).squeeze(), :].index.tolist())
    df[metric] = df[metric].astype(float)
    
    return df


runs = scrub_strings("test/target/rmse/dataloader_idx_1", runs)
runs = scrub_strings("test/target/score/dataloader_idx_1", runs)
```

```python
pivoted_runs = runs.pivot_table(
    index=["adaption_mode", "backbone", "approach"],
    columns=["dataset", "source", "target"], 
    values=["test/target/rmse/dataloader_idx_1"],
)
pivoted_runs
```

```python
score_pivot = runs.pivot_table(
    index=["adaption_mode", "backbone", "approach"],
    columns=["dataset", "source", "target"], 
    values=["test/target/score/dataloader_idx_1"],
)
score_pivot
```

```python
pivoted_runs.to_csv("rmse.csv")
score_pivot.to_csv("score.csv")
runs.to_csv("runs.csv")
```
